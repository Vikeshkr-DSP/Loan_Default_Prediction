### Defining a function to find set of features having correlation greater than the threshold

```python
def correlation(dataset,thrsld):
    col_corr=set()
    corr_matrix=dataset.corr()
    for col1 in range(len(corr_matrix.columns)):
        for col2 in range(col1):
            if abs(corr_matrix.iloc[col1,col2])>thrsld:
                colname=corr_matrix.columns[col1]
                col_corr.add(colname)
    return col_corr
```

```python
featr_tb_rmvd=correlation(X_train,0.85)
featr_tb_rmvd
```

Keeping multiple features which are highly correlated is of no use because the model will learn the same thing from different features and keeping multiple features which correlated will take more computational time. So, its better to keep only one such feature and remove other highly correlated feature.

```python
X_train.drop(featr_tb_rmvd,axis=1,inplace=True)   # Removing highly correlated features
```

## Feature Importance for Random Forest

```python
# defining a function to plot feature importance

def plot_feature_importance(model,algo):
    
    # finding the important features
    DF = pd.DataFrame({'Feature_Name':list(X_train),'Feature_importance':model.feature_importances_})
    DF.sort_values(by='Feature_importance', ascending=False,inplace=True)
    
    # plotting the feature importance
    plt.figure(figsize=(17,15))
    sns.barplot(x=DF['Feature_importance'], y=DF['Feature_Name'])
    plt.xlabel('Feature Importance')
    plt.ylabel('Feature')
    plt.title(algo+"'s Feature Importance")
```

```python
plot_feature_importance(rdf,'Random Forest')
```

## Finding & Removing Insignificant Feature for Random Forest

```python
# defining a function to find feature having less importance than the thresold

def features_tb_removed(model,thrsld):
    
    # finding feature importance
    DF = pd.DataFrame({'Feature_Name':list(X_train),'Feature_importance':model.feature_importances_})

    # finding features having less importance than thresold    
    feature=set()                                # variable to append feature
    for row_no in range(DF.shape[0]):
        if DF.iloc[row_no,1] < thrsld:           # conditional line to find insignificant feature
            feature.add(DF.iloc[row_no,0])       # appendong insignificant features to variable feature
    return feature
```

## Conclusion
Three algorithm used for training are Random Forest, XGBoost and Logistic Regression. This problem is imbalanced class problem but all the algorithms are good in differentiating between the classes without under or over sanpling which can be seen in the AUC score. All these three algorithms performance is explained below.
### Random Forest:-
Random with default parameters performed very well (actually more than expected) with an AUC score of 0.9968. Also by default, there were 43 features bur some of them were not helping the model to learn the pattern and hence those features were removed. After removing insignificant features, there were 33 features on which the model can be trained. Training Random Forest with default parameters on significant features have an AUC score of 0.9966 which is good when we also consider the number of features.
### XGBoost:-
XGBoost with it's default parameters worked better than Random Forest and gave an AUC score of 0.9999. But XGBoost was learning from very few features and thus most them were removed. After removing insgnificant features, there were only 11 features and gave an AUC score of 0.9992 on validation data which is very significant considering no of features. I went for HyperParameter Tuning for (XGB) using GPU on significant features and got AUC score of 0.9987 but the result obtained was best out of all methods tried before and thus XGBoost model after HyperParameter Tuning is being selected as our model for web app.
### Logistic Regression:-
Logistic Regression worked very well after rescaling all the features and gave an AUC score of 0.9689 using all the features. Also Hyper Tuning Logistic Regression using all the features will computationaly costly and will take longer for doing predictions. Hence, we stopped here and decided to go with XGBoost with HyperParameter Tuning.
